# Importando bibliotecas

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Importando dados

```python
e = pd.read_csv('energia.dat', delim_whitespace=True, names=['t', 'e'])
theta = pd.read_csv('oscilar.dat', delim_whitespace=True, names=['t', 'theta'])
```

# Gr√°ficos

```python
# Energia
f1, ax1 = plt.subplots()
ax1.plot(e['t'], e['e'], label='E(t)')
ax1.set_ylabel('Energia (J)')
ax1.set_xlabel('Tempo (s)')
ax1.grid()
ax1.legend()
```

```python
# Theta(t)
f2, ax2 = plt.subplots()
ax2.plot(theta['t'], theta['theta'], label='$\\theta(t)$')
ax2.set_ylabel('$\\theta$ (rad)')
ax2.set_xlabel('Tempo (s)')
ax2.grid()
ax2.legend()
f2.savefig('./figura-2.pdf', dpi=300, bbox_inches='tight')
```
